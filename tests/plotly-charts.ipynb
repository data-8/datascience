## Working with Returned Figures

```python
# Example 1: Taking scatter plot, changing colors of fit lines
fig = more_cats.select(['weight', 'displ', 'hp']).scatter('weight', fit_line=True, overlay=False, show=False)

# Marker argument in update_traces defines what the new attribute should be defined to, selector finds the specific trace
fig.update_traces(marker={"color": "red"}, selector={"name": "displ best fit line"})
fig.update_traces(marker={"color": "green"}, selector={"name": "hp best fit line"})

fig.show()
```

```python
# Example 2: Updating opacity, hover text, and name of bar charts
fig = more_cats.group('cyl', max).barh('cyl', [1, 2, 3], show=False)

# Update_traces without a selector will update all traces
fig.update_traces(marker={"opacity": 0.4})

# To update fields that update_traces does not allow us to modify, we can manually modify the dictionary values
for trace in fig["data"]:
    trace["hovertemplate"] = "new hovertemplate: %{x}, %{customdata}"
    trace["name"] = "new name: " + trace["name"]
    
fig.show()
```

```python
# Example 3: Creating background with shapes, adding additional trace
import plotly.graph_objects as go

fig = more_cats.plot('weight', ['displ', 'hp'], show=False)

# Adding rectangle shape in the background to highlight a specific section
fig.update_layout(
    shapes=[dict(
        type="rect",
        xref="x",     # Units are the same as the x coordinates
        yref="paper", # y endpoints are in [0, 1] that scale with the plot paper
        x0=3500,
        x1=4000,
        y0=0,
        y1=1, 
        line_width=0,
        fillcolor="LightSeaGreen",
        opacity=0.3,
        layer = "below"
    )]
)

# Adding an additional scatter plot trace
x = np.linspace(1500, 5000, 100)
fig.add_trace(
    go.Scatter(
        x = x, 
        y = 40 * np.sin(x) + 0.1 * x,
        mode = "markers",
        marker_color = "red"
    )
)

fig.show()
```
